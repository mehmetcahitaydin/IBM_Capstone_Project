## Demo: How to work with database


Download database file.

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
import seaborn as sns
import matplotlib.pyplot as plt
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY="""
SELECT ConvertedComp FROM master

"""
converted=pd.read_sql_query(QUERY,conn)

plt.hist(converted, bins=50)
plt.gca().set(title='ConvertedComp', ylabel='Frequency');
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY="""
SELECT Age FROM master

"""
age=pd.read_sql_query(QUERY,conn)

age.plot(kind="box",vert=False,figsize=(8,4))
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY="""
SELECT Age,WorkWeekHrs FROM master

"""
age_Work=pd.read_sql_query(QUERY,conn)

age_Work.plot(kind="scatter",x="Age",y="WorkWeekHrs",title="Scatter Plot")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY="""
SELECT Age,WorkWeekHrs,CodeRevHrs FROM master

"""
age_work_cod=pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=age_work_cod, x="WorkWeekHrs", y="CodeRevHrs", size="Age", legend=False, sizes=(20, 1600))
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here



QUERY = """
SELECT COUNT(DatabaseDesireNextYear) as count,DatabaseDesireNextYear FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
Order by count DESC
LIMIT 5
"""

database_desire=pd.read_sql_query(QUERY,conn)

database_desire['count'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%',                                 
                            shadow=True,           
                            labels=database_desire['DatabaseDesireNextYear'],title="Top 5 databases that respondents wish to learn next year")
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age 
FROM master
Where Age Between 30 And 35

"""
yeter=pd.read_sql_query(QUERY,conn)

grafik=yeter.groupby(["Age"])[["WorkWeekHrs","CodeRevHrs"]].median()

grafik.plot(kind='bar', figsize=(10, 6), stacked=True)

plt.xlabel('Age') # add to x-label to the plot
plt.ylabel('Hours') # add y-label to the plot
plt.title('Median Hours by Age') # add title to the plot

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp,Age
FROM master
Where Age Between 45 and 60
Order By Age 
"""
convertedComp=pd.read_sql_query(QUERY,conn)

LineChart=convertedComp.groupby(["Age"])["ConvertedComp"].median()

LineChart.plot(kind='line', figsize=(10, 6),ylabel="Median of ConvertedComp")
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch,count(*)
FROM master
group by MainBranch

"""
MainBranch=pd.read_sql_query(QUERY,conn)

MainBranch.plot(kind="barh",figsize=(10,6))
```
